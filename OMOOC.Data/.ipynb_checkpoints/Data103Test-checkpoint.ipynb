```python
import pandas_datareader
from pandas_datareader import data
import pandas as pd

import datetime
start = datetime.datetime(2015,10,1)
end = datetime.datetime(2016,10,1)
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.graphics.api as smg
import patsy
```

```python
pingan=data.DataReader("000001.SZ","yahoo",start,end) # 000001
```

```python
sse = data.DataReader("000001.SS", "yahoo", start, end)
```

```python
data=pd.concat([sse["Close"],pingan["Close"]],axis=1).dropna()
```

```python
data.columns=["x",'y']
```

```python
data.head()
```

```python
data["xx"] = data["x"].diff()/data["x"].shift(1)
```

```python
data["yy"] = data["y"].diff()/data["y"].shift(1)
```

```python
data=data.dropna()
```

```python
data.head()
```

```python
data1 = data.drop(["y","x"],axis=1)
```

```python
data1.head()
```

```python
from scipy import stats
import seaborn as sns
```

```python
model1 = smf.ols("yy ~ -1+xx",data1)
result1 = model1.fit()
result1.summary()
```

```python
z, p = stats.normaltest(result1.resid.values)
```

```python
z,p
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(8,4))

x= np.linspace(-4,4,100)
y = x
ax.plot(x,y)
smg.qqplot(result1.resid, line="q",ax=ax)

fig.tight_layout()
```

```python
model = smf.ols("y ~ x",data)
result = model.fit()
result.summary()
```
